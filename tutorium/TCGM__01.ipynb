# Andere MÃ¶glichkeit einen Sinuston zu generieren 

```python
sr = 44100
wts = sr * 2
wt = np.zeros(wts)
freq = 440
for i in range(wts):
    wt[i] = np.sin(2 * np.pi * (float(i)/float(wts)) * freq)
display(Audio(wt, rate = sr))
```

### Funktion zur Generierung eines Sinustons

```python
def sine(index,wts,freq,freqscale,ampscale):
    return np.sin(2 * np.pi * (float(index)/float(wts)) * freq * freqscale) * ampscale
```

# Generierung eines Signals mit mehreren ObertÃ¶nen

#### "Formel": 
$ sin(2  \pi  \frac{float(index)} {float(wts)}  freq)  ampscale + ... $


```python
ampscale = 1 / 4 #4 ObertÃ¶ne
freq = 440 # Grundfrequenz
for i in range (wts):
    wt[i] = sine(i, wts, freq, 1, ampscale) + sine(i, wts, freq, 2, ampscale) + sine(i, wts, freq, 3, ampscale) + sine(i, wts, freq, 4, ampscale)
    
display(Audio(wt, rate = sr))
plt.plot(wt)
plt.ylim(-1,1)
plt.xlim(0,wts /freq)
```

# Frequenzmodulation (FM)
[FM doc](https://en.wikipedia.org/wiki/Frequency_modulation_synthesis) \
[Markdown doc](https://rpruim.github.io/s341/S19/from-class/MathinRmd.html)

$FM(t) = A \text{ sin}(\omega_{c} t + \beta (sin \omega_{m} t))  $ 
>$\omega = 2 \pi f$ 


```python
wt2 = np.zeros(sr * 2)
carrierfreq = 220
moddepth = 880
modfreq = 440
for i in range(wts):
    wt2[i] = 0.5 * np.sin(2 * np.pi * (float(i) / float(wts)) * carrierfreq + moddepth/modfreq * np.sin(2 * np.pi * (float(i) / float(wts)) * modfreq))
plt.plot(wt2)
plt.ylim(-1,1)
plt.xlim(0,wts / carrierfreq )                                                                                        
display(Audio(wt2, rate = sr))
```
