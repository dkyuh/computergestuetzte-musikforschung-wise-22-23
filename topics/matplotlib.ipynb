# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## `plot`

$f(x) = x$

```python
x = np.arange(-10, 10, 0.5)
y = x

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-10, 10, 0.5)
y = (x ** 3) - 50

plt.figure(figsize=(10, 10))
plt.plot(x, y, 'o-')
# plt.xlim(-5, 5)
plt.show()
```
